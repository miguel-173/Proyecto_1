Exploración:  Entrega el estrato más común en una carrera específica

```python
from matplotlib import pyplot as plt
from collections import Counter

def moda_estrato_carrera(nombre_carrera):
  l = []
  for i in range(len(df['ESTU_PRGM_ACADEMICO'])):
    if df['ESTU_PRGM_ACADEMICO'][i] == 'INGENIERIA ELECTRICA':
      l.append(df['FAMI_ESTRATOVIVIENDA'][i])
  contador = Counter(l)
  moda = contador.most_common(1)[0][0]
  b = print("El estrato más común en los estudiantes de ", nombre_carrera, "es ", moda, ".")
  return b

moda_estrato_carrera("INGENIERIA ELECTRICA")



```

Exploración:  Entrega el número de estudiantes por cada carrera.

```python

prgm = df['ESTU_PRGM_ACADEMICO'].tolist()

frecuencia = {}

for prog in prgm:

  if prog in frecuencia:
    frecuencia[prog] += 1
  else:
    frecuencia[prog] = 1

for prog, frec in frecuencia.items():
  print( f"{prog} : {frec} estudiantes.")

```
