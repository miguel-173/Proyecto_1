##LIMPIEZA DE DATOS

*   Valores faltantes por columna

```python
# Verificar valores faltantes
print("Valores faltantes por columna:")
print(df.isnull().sum())

# Verificar duplicados
print("\nNúmero de filas duplicadas:", df.duplicated().sum())

# Eliminar filas con valores faltantes
df_sin_nulos = df.dropna()

# Eliminar filas duplicadas
df_sin_duplicados = df.drop_duplicates()

# Verificar la forma de los datos después de la limpieza
print("\nForma de los datos después de eliminar filas con valores faltantes y duplicados:")
print("Filas:", df_sin_nulos.shape[0], "Columnas:", df_sin_nulos.shape[1])
```

##CODIFICACIÓN

*   Codificación One - Hot
*   Codificación de etiquetas

```python
# Codificación One-Hot
df_codificado = pd.get_dummies(df)

# Mostrar las primeras filas del DataFrame codificado
print(df_codificado.head())
```

```python
from sklearn.preprocessing import LabelEncoder

# Crear un objeto LabelEncoder
le = LabelEncoder()

# Codificar variables categóricas
df_codificado_etiquetas = df.copy()  # Hacemos una copia del DataFrame original
for columna in df_codificado_etiquetas.columns:
    if df_codificado_etiquetas[columna].dtype == 'object':
        df_codificado_etiquetas[columna] = le.fit_transform(df_codificado_etiquetas[columna])

# Mostrar las primeras filas del DataFrame con codificación de etiquetas
print(df_codificado_etiquetas.head())
```

##NORMALIZACIÓN DE LOS DATOS

*   MIN-MAX SCALER


```python
#Normalización de los datos

from sklearn.preprocessing import MinMaxScaler

# Creamos un objeto MinMaxScaler
scaler = MinMaxScaler()

# Normalizamos las características
df_normalizado_minmax = df_codificado_etiquetas.copy()  # Utilizamos el DataFrame codificado previamente por codificación
df_normalizado_minmax[df_normalizado_minmax.columns] = scaler.fit_transform(df_normalizado_minmax[df_normalizado_minmax.columns])

# Mostramos las primeras filas del DataFrame normalizado con Min-Max Scaling
print(df_normalizado_minmax.head())
```
